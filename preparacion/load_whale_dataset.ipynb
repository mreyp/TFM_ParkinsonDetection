```python
from scipy.io import arff
import pandas as pd

label_name = 'class'
labels = ['RightWhale', 'NoWhale']

data_train, meta_train = arff.loadarff('TFM_MartaRey/datos/whale_dataset/RightWhaleCalls_TRAIN.arff')
print(meta_train)
df_train = pd.DataFrame(data_train)
print(df_train[label_name].value_counts())

df = df_train.rename(columns={label_name: 'label'})
df['label'] = df['label'].str.decode("utf-8")
for i, one_label in enumerate(labels):
    df.loc[df.label == one_label, 'label'] = int(i)
print(df['label'].value_counts())
# print(df['label'].value_counts(normalize=True))

print(len(df))
df_train_7 = df.sample(frac=0.7, random_state=1)
print(len(df_train_7))
df_val_3 = df[~df.index.isin(df_train_7.index)]
print(len(df_val_3))
print(len(df_train_7) + len(df_val_3))

df_train_7.to_csv('TFM_MartaRey/datos/whale_dataset/RightWhaleCalls' + '_train.csv', index=False)
df_val_3.to_csv('TFM_MartaRey/datos/whale_dataset/RightWhaleCalls' + '_val.csv', index=False)


data_test, meta_test = arff.loadarff('TFM_MartaRey/datos/whale_dataset/RightWhaleCalls_TEST.arff')
df_test = pd.DataFrame(data_test)

df = df_test.rename(columns={label_name: 'label'})
df['label'] = df['label'].str.decode("utf-8")
for i, one_label in enumerate(labels):
    df.loc[df.label == one_label, 'label'] = int(i)
print(df['label'].value_counts())
# print(df['label'].value_counts(normalize=True))

print(len(df))
df.to_csv('TFM_MartaRey/datos/whale_dataset/RightWhaleCalls' + '_test.csv', index=False)
```

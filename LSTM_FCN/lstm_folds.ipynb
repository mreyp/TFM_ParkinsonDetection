```python
import numpy as np
import os
import shutil
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import roc_auc_score, confusion_matrix, roc_curve, auc

original_directory = '/media/beegfs/home/v890/v890699/'

os.chdir('/media/beegfs/home/v890/v890699/TFM_MartaRey/LSTM_FCN')
from lstm_fcn_pytorch.model_1_3_folds import Model

saving_directory = 'results_parkinson_folds_real_1206_6'

if os.path.exists(saving_directory):
    shutil.rmtree(saving_directory)
    print(f"Carpeta '{saving_directory}' eliminada.")
os.makedirs(saving_directory)
print(f"Carpeta '{saving_directory}' creada.")

all_predictions = []
all_true_labels = []
all_metrics = []
tprs = []
mean_fpr = np.linspace(0, 1, 100)
folds = 5
for fold_index in range(1, folds+1): 
    print(f"PROCESSING FOLD {fold_index}")
    os.chdir(original_directory)

    fold_files_directory = 'TFM_MartaRey/datos/sets/folds_5/files_real_40_1e5_N'
    dataset_id = '40_1e5_N'
    
    # Load data
    x_train = np.load(os.path.join(fold_files_directory, f'X_train_{dataset_id}_fold_{fold_index}.npy'))
    y_train = np.load(os.path.join(fold_files_directory, f'y_train_{dataset_id}_fold_{fold_index}.npy'))
    x_val = np.load(os.path.join(fold_files_directory, f'X_val_{dataset_id}_fold_{fold_index}.npy'))
    y_val = np.load(os.path.join(fold_files_directory, f'y_val_{dataset_id}_fold_{fold_index}.npy'))
    x_test_real = np.load(os.path.join(fold_files_directory, f'X_test_{dataset_id}_fold_{fold_index}.npy'))
    y_test_real = np.load(os.path.join(fold_files_directory, f'y_test_{dataset_id}_fold_{fold_index}.npy'))

    print('X train', x_train.shape)
    print('y train', y_train.shape)
    print('X val', x_val.shape)
    print('y val', y_val.shape)
    print('X test real', x_test_real.shape)
    print('y test real', y_test_real.shape)

    os.chdir('TFM_MartaRey/LSTM_FCN')

    # Model
    model = Model(
        x=x_train,
        y=y_train,
        units=[4,4],
        filters=[64, 32, 64],
        kernel_sizes=[8, 5, 3],
        dropout=0.9,
        dropout2=0.0,
        val_data=(x_val, y_val)
    )

    # Training
    model.fit(
        learning_rate=1e-4,
        batch_size=256,
        epochs=500, 
        verbose=True,
        saving_directory = saving_directory,
        fold_index=fold_index
    )

    # Evaluation
    yhat_test_r_labels, yhat_test_r = model.predict(x_test_real)
    print('Final test real for fold {}:'.format(fold_index))
    metrics = model.evaluate(yhat_test_r_labels, yhat_test_r, y_test_real, saving_directory, 
                                fold_index, class_names=['Control', 'Parkinson'])

    all_predictions.extend(yhat_test_r_labels)
    all_true_labels.extend(y_test_real)
    accuracy, precision, recall, f1, auc_score, fpr, tpr, thresholds = metrics
    all_metrics.append([accuracy, precision, recall, f1, auc_score])

    # Interpolation of TPR for common FPR
    interp_tpr = np.interp(mean_fpr, fpr, tpr)
    interp_tpr[0] = 0.0
    tprs.append(interp_tpr)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

os.chdir('/media/beegfs/home/v890/v890699/TFM_MartaRey/LSTM_FCN')

# Mean and Std
mean_metrics = np.mean(all_metrics, axis=0)
std_metrics= np.std(all_metrics, axis=0)
# Pearson Variation Coefficient
cv_metrics = (std_metrics / mean_metrics)

print(f"Mean Metrics - Accuracy: {mean_metrics[0]:.2f}, Precision: {mean_metrics[1]:.2f}, Recall: {mean_metrics[2]:.2f}, F1 Score: {mean_metrics[3]:.2f}, AUC: {mean_metrics[4]:.2f}")
print(f"Variation Coefficients - Accuracy: {cv_metrics[0]:.2f}, Precision: {cv_metrics[1]:.2f}, Recall: {cv_metrics[2]:.2f}, F1 Score: {cv_metrics[3]:.2f}, AUC: {cv_metrics[4]:.2f}")

data = {
    'Metrics': ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC'],
    'Values': mean_metrics,
    'CV Metrics': ['Accuracy CV', 'Precision CV', 'Recall CV', 'F1 Score CV', 'AUC CV'],
    'CV Values': cv_metrics
}

metrics_df = pd.DataFrame(data)
metrics_df = metrics_df.T
metrics_df.columns = metrics_df.iloc[0]
metrics_df = metrics_df.drop(metrics_df.index[0])
csv_file_path = saving_directory + f'/mean_test_metrics.csv'
metrics_df.to_csv(csv_file_path, index=False)
##-----------------------------------------------------------------CONFUSION MATRIX

class_names=['Control', 'Parkinson']
cm_all = confusion_matrix(all_true_labels, all_predictions)

print(classification_report(all_true_labels, all_predictions, target_names=class_names))

plt.figure(figsize=(5,4))
sns.heatmap(cm_all, annot=True, fmt="d", cmap="Blues", cbar=False, xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Accumulated Confusion Matrix')
plt.savefig(saving_directory + '/accumulated_confusion_matrix.png', dpi=300)
plt.show()

##-----------------------------------------------------------------ROC CURVE
### auc_all vs mean_auc

# Calcular la media y la desviación estándar de los TPRs
mean_tpr = np.mean(tprs, axis=0)
mean_tpr[-1] = 1.0

fig, ax = plt.subplots(figsize=(5,4))
ax.plot(mean_fpr, mean_tpr, color='b',
        label='Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_metrics[4], std_metrics[4]),
        lw=1)
ax.plot([0, 1], [0, 1], color='black', lw=1, linestyle='--')
ax.grid()

# Añadir la variabilidad de la curva ROC (desviación estándar)
std_tpr = np.std(tprs, axis=0)
tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
ax.fill_between(mean_fpr, tprs_lower, tprs_upper, color='grey', alpha=0.2,
                label='$\pm$ 1 std. dev.')

# Añadir detalles finales al gráfico
ax.set(xlabel='False Positive Rate', ylabel='True Positive Rate', title='Mean ROC Curve')
ax.legend(loc='lower right')
plt.savefig(saving_directory + '/mean_roc_curve.png', dpi=300)
plt.show()
```
